# Russian Athletes

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv(
    '../data/marathon.csv',
    sep=',',
    header=None,
    names=[
        'Rank',
        'Time',
        'Name',
        'Country',
        'Date of Birth',
        'Place',
        'City',
        'Date',
        'Gender',
    ]
)
```

```python
df = df.sort_values('Time')
```

Show all Russian atheletes

```python
rus_df = df[df['Country'] == 'RUS']
rus_df
```

Get unique names

```python
rus_df.groupby(rus_df["Name"])\
    .size()\
    .sort_values(ascending=False)
```
