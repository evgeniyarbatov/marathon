```python
import pandas as pd
```

```python
df = pd.read_csv(
    '../data/marathon.csv',
    sep=',',
    header=None,
    names=[
        'Rank',
        'Time',
        'Name',
        'Country',
        'Date of Birth',
        'Place',
        'City',
        'Date',
        'Gender',
    ]
)
```

```python
df.size
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'], errors='coerce')
```

```python
df['Pace'] = pd.to_timedelta(df['Time']).dt.total_seconds() / (60 * 42.195)
df['Age'] = round((df['Date'] - df['Date of Birth']).dt.days / 365)
```

```python
df['Time Delta'] = pd.to_timedelta(df['Time']).diff().dt.total_seconds()
```

```python
df = df[df['Age'] >= 18]
```
